```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'http://chubbygrub.com/'
```

```python
response = requests.get(url)
response.status_code
```

```python
soup = BeautifulSoup(response.content, 'lxml')
```

```python
def get_res(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'lxml')
    return soup
```

```python
chubby = []
table = soup.find('div', {'class': 'restaurant-buttons'})
restaurant_list = []
for row in table.find_all('a'):
    restaurant = row.text
    restaurant_url = url + row.attrs['href']
    soup_res = get_res(restaurant_url)
    for row in soup_res.find('tbody').find_all('tr'):
        food = {}
        td = row.find_all('td')
        food['restaurant'] = restaurant
        food['name'] = td[0].text
        food['categories'] = td[1].text.strip()
        food['calories'] = td[2].text
        food['fat'] = td[3].text
        food['carbs'] = td[4].text
        chubby.append(food)
```

```python
df = pd.DataFrame(chubby)
df
```
